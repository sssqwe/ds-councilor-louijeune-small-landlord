```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('income-restricted-inventory-2021-address-added.csv')
```

```python
data_address =  data.apply(lambda x: x['address_candidates'].strip('][').split("', '"),axis = 1)
                                                             
```

```python
useable_addresses = []
for i, tf in enumerate([len(x) == 1 for x in data_address]):
    if tf:
        useable_addresses.append(data_address[i][0].strip("'").strip(', United States').upper())
    
```

```python
useable_addresses
```

```python
assesor = pd.read_csv('Property_Assessment_FY2022.csv')
```

```python
assesor
```

```python
assesor['lookup'] = assesor.apply(lambda x : x['ST_NUM'].strip() + ' ' + x['ST_NAME'] + ', ' + x['CITY'] + ' 0' + str(x['ZIPCODE']), axis = 1)
```

```python
assesor.loc[assesor['lookup'].isin(useable_addresses)]
```
